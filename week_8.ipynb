```python
##Step 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

```python
##step2 Step 2. Import the dataset from this address.
!wget https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv
```

```python
##Step 3. Assign it to a variable called chipo.
chipo = pd.read_table('chipotle.tsv')
chipo.dropna()
```

```python
##Step 4. See the first 10 entries
chipo.head(10)
```

```python
##Step 5. Create a histogram of the top 5 items bought
item_counts = chipo['item_name'].value_counts()
top_items = item_counts.head(5)
top_items.plot(kind='bar')
plt.title('Top 5 Items Bought')
plt.show()
```

```python
#Step 6. Create a scatterplot with the number of items orderered per order price
chipo.plot(kind='scatter', x='item_price', y = 'quantity')
plt.show()
```

```python
#Step 7. BONUS: Create a question and a graph to answer your own question.
item_counts = chipo['item_name'].value_counts().sort_values(ascending=False)
top_items = item_counts.head(10)
plt.barh(top_items.index, top_items.values)
plt.show()
```
